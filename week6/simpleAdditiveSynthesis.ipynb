```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display
```

```python
samplingRate = 44100
frequency = 100

# generate list of harmonics
#harmonics = np.arange(1,1,1) # sine wave
#harmonics = np.arange(2,11,1) # all harmonics
harmonics = np.arange(2,11,2) #odd harmonics

# create a time series from 0 to 1 the inputted length    
timeSeries = np.linspace(0, 1, samplingRate)
    
# create a cosine wave for the fundamental    
signal = np.cos(2.0 * np.pi * frequency * timeSeries)

print(frequency)
# add harmonics based on the those specified in the variable 'harmonics'
for harm in harmonics:
    signal = (signal + np.cos(2.0 * np.pi * (frequency * harm) * timeSeries))/2

# plot signal
plt.plot(timeSeries,signal)
plt.xlabel('Time (Sec)')
plt.ylabel('Amplitude')
plt.title('Additive Synthesis')
plt.plot
    
# play complex signal
IPython.display.Audio(data=signal, rate=samplingRate)
```

```python
def simpleAdditiveSynthesis(samplingRate,frequency,lowestHarm,highestHarm,seqHarm):

    # generate list of harmonics
    harmonics = np.arange(lowestHarm,highestHarm,seqHarm)
    
    # create a time series from 0 to 1 the inputted length    
    timeSeries = np.linspace(0, 1, samplingRate)
    
    # create a cosine wave for the fundamental    
    signal = np.cos(2.0 * np.pi * frequency * timeSeries)

    # add harmonics based on the those specified in the variable 'harmonics'
    for harm in harmonics:
        signal = (signal + np.cos(2.0 * np.pi * frequency * harm * timeSeries))/2     
        
    return signal
```

```python
samplingRate = 44100
frequency = 100
lowestHarm = 1
highestHarm = 11 #plus one
seqHarm = 2

signal1 = simpleAdditiveSynthesis(samplingRate,frequency,lowestHarm,highestHarm,seqHarm)
IPython.display.Audio(data=signal1, rate=samplingRate)
```
