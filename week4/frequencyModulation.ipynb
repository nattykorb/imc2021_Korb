```python
import numpy as np
import matplotlib.pylab as plt
import librosa.display
import IPython.display
import librosa
```

```python
def subplots(plotPos,title,data,color,ylabel,xlabel):
    plt.subplot(plotPos[0],plotPos[1],plotPos[2])
    plt.title(title)
    plt.plot(data,color)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
```

```python
def frequencyModulation(carrierFreq,modulatorFreq,modulatorAmp,modIndex,length):
    timeSeries = np.linspace(0, 1, length)
    
    # remember AM modulation
    #modulator = modulatorAmp*np.cos(2*np.pi*modulatorFreq*timeSeries)
    #product = (1+modIndex*modulator)*carrier
    
    # note different for FM modulation
    carrier = np.sin(2.0 * np.pi * carrierFreq * timeSeries)
    modulator = np.sin(2.0 * np.pi * modulatorFreq * timeSeries) * modIndex
    product = np.sin(2. * np.pi * (carrierFreq * timeSeries + modulator))
    
    subplots([3,1,1],'Frequency Modulation',modulator,'g','Amplitude','Modulator signal')
    subplots([3,1,2],'Frequency Modulation',carrier,'r','Amplitude','Carrier signal')
    subplots([3,1,3],'Frequency Modulation',product,'purple','Amplitude','AM signal')
    
    plt.subplots_adjust(hspace=1)
    plt.rc('font', size=15)
    fig = plt.gcf()
    fig.set_size_inches(16, 9)
    
    return product
```

```python
modSignal=frequencyModulation(440,20,0.5,1,44100)
IPython.display.Audio(data=modSignal, rate=44100)
```
